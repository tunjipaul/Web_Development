## Introduction
Today we'll make our web pages interactive! Forms allow users to input data, media elements make pages engaging, and proper navigation helps users move around your site. These elements transform static pages into dynamic, user-friendly experiences.

```python
%%html
<div style="background-color: #f9f9f9; padding: 20px; border-radius: 10px; max-width: 500px;">
    <h3 style="color: #2c3e50;">Basic Contact Form</h3>
    
    <form>
        <div style="margin-bottom: 15px;">
            <label for="fullname" style="display: block; margin-bottom: 5px; font-weight: bold; color: black;">Full Name:</label>
            <input type="text" id="fullname" name="fullname" 
                   style="width: 90%; padding: 8px; border: 2px solid #ddd; border-radius: 5px;" 
                   placeholder="Enter your full name">
        </div>
        
        <div style="margin-bottom: 15px;">
            <label for="email" style="display: block; margin-bottom: 5px; font-weight: bold; color: black;">Email:</label>
            <input type="email" id="email" name="email" 
                   style="width: 90%; padding: 8px; border: 2px solid #ddd; border-radius: 5px;" 
                   placeholder="your.email@example.com">
        </div>
        
        <div style="margin-bottom: 15px;">
            <label for="message" style="display: block; margin-bottom: 5px; font-weight: bold; color: black;">Message:</label>
            <textarea id="message" name="message" rows="4" 
                      style="width: 90%; padding: 8px; border: 2px solid #ddd; border-radius: 5px;" 
                      placeholder="Write your message here..."></textarea>
        </div>
        
        <button type="submit" 
                style="background-color: #3498db; color: white; padding: 12px 20px; border: none; border-radius: 5px; cursor: pointer; font-size: 16px;">
            Send Message
        </button>
    </form>
</div>
```

## Links and Navigation
Links connect web pages together and enable navigation.

```python
%%html
<div style="background-color: #f0f8ff; padding: 20px; border-radius: 10px; color: black;">
    <h2 style="color: #2980b9;">HTML Links</h2>
    
    <div style="margin-bottom: 25px;">
        <h4>External Links</h4>
        <p><a href="https://developer.mozilla.org/en-US/docs/Web/HTML" target="_blank" 
              style="color: #3498db; text-decoration: none; border-bottom: 2px solid #3498db;">
            MDN HTML Documentation
        </a></p>
        <p><a href="https://www.w3schools.com/html/" target="_blank"
              style="color: #e74c3c; text-decoration: none; border-bottom: 2px solid #e74c3c;">
            W3Schools HTML Tutorial
        </a></p>
    </div>
    
    <div style="margin-bottom: 25px;">
        <h4>Email and Phone Links</h4>
        <p><a href="mailto:student@example.com" 
              style="color: #27ae60; text-decoration: none;">
            📧 Send Email
        </a></p>
        <p><a href="tel:+1234567890" 
              style="color: #f39c12; text-decoration: none;">
            📞 Call Us
        </a></p>
    </div>
    
    <div style="margin-bottom: 25px;">
        <h4>Internal Page Links (Anchors)</h4>
        <p><a href="#top" style="color: #8e44ad; text-decoration: none;">Go to Top</a></p>
        <p><a href="#summary" style="color: #8e44ad; text-decoration: none;">Jump to Summary</a></p>
    </div>
    
    <div>
        <h4>Download Link</h4>
        <p><a href="data:text/plain;charset=utf-8,Hello%20World!" 
              download="sample.txt"
              style="background-color: #34495e; color: white; padding: 8px 16px; text-decoration: none; border-radius: 4px;">
            Download Sample File
        </a></p>
    </div>
</div>
```

**Link Attributes:**
- `href` - Link destination
- `target="_blank"` - Open in new tab
- `download` - Force download
- `title` - Tooltip text

**Table Elements:**
- `<table>` - Table container
- `<thead>` - Table header section
- `<tbody>` - Table body section
- `<tfoot>` - Table footer section
- `<tr>` - Table row
- `<th>` - Header cell
- `<td>` - Data cell

### 📘 Simple Colspan Example

```python
%%html
<div style="background-color: #f8f9fa; padding: 20px; border-radius: 10px;">
    <h3 style="color: #495057;">Simple Colspan Example</h3>
    
    <table style="width: 60%; border-collapse: collapse; background-color: white; box-shadow: 0 2px 6px rgba(0,0,0,0.1); color:black">
        <thead>
            <tr style="background-color: #007bff; color: white;">
                <th style="padding: 10px; border: 1px solid #ddd;">ID</th>
                <th colspan="2" style="padding: 10px; border: 1px solid #ddd; text-align:center">Student Info</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td style="padding: 10px; border: 1px solid #ddd;">1</td>
                <td style="padding: 10px; border: 1px solid #ddd;">John Doe</td>
                <td style="padding: 10px; border: 1px solid #ddd;">20</td>
            </tr>
            <tr style="background-color: #f8f9fa;">
                <td style="padding: 10px; border: 1px solid #ddd;">2</td>
                <td style="padding: 10px; border: 1px solid #ddd;">Jane Smith</td>
                <td style="padding: 10px; border: 1px solid #ddd;">22</td>
            </tr>
        </tbody>
    </table>
</div>
```

In the header row, **"Student Info"** uses `colspan="2"`, which means it spans across **two columns** (Name + Age).

## Audio and Video Elements
HTML5 provides native support for multimedia content.

```python
%%html
<div style="background-color: #fff5f5; padding: 20px; border-radius: 10px; color:black;">
    <h3 style="color: #e74c3c;">Multimedia Elements</h3>
    
    <div style="margin-bottom: 40px;">
        <h4>Audio Element</h4>
        <audio controls style="width: 100%;">
            <source src="https://www.soundjay.com/misc/sounds/bell-ringing-05.wav" type="audio/wav">
        </audio>
    </div>
    
    <div style="margin-bottom: 40px;">
        <h4>Video Element</h4>
        <video controls width="100%" style="max-width: 500px; border-radius: 8px;">
            <source src="https://www.w3schools.com/html/mov_bbb.mp4" type="video/mp4">
        </video>
        <p style="font-size: 14px; color: #666; margin-top: 8px;">
            Sample video from W3Schools
        </p>
    </div>
    
    <div>
        <h4>YouTube Embed (iframe)</h4>
        <iframe width="560" height="315"
            src="https://www.youtube.com/embed/UB1O30fR-EE"
            title="HTML Crash Course"
            frameborder="0"
            allowfullscreen>
        </iframe>
    </div>
</div>
```

### 🎥 Media & YouTube Embed Reference

**Media Attributes (for `<video>` / `<audio>`):**
- `controls` – Show playback controls  
- `autoplay` – Start playing automatically  
- `loop` – Repeat playback  
- `muted` – Start muted  

**YouTube Embed (using `<iframe>`):**
- `src` – The YouTube embed link (`https://www.youtube.com/embed/VIDEO_ID`)  
- `title` – Description for accessibility  
- `width` & `height` – Size of the video frame  
- `allowfullscreen` – Lets video open in full screen
